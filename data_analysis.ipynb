```python
# Removes outliers from the specified numerical columns in a dataframe using the IQR method.

def remove_outliers(df, columns):
    """
    Removes outliers from the specified numerical columns in a dataframe using the IQR method.
    
    Parameters:
        df (DataFrame): The dataset containing the data.
        columns (list): List of numerical columns to remove outliers from.
    
    Returns:
        DataFrame: A dataframe with outliers removed.
    """
    for col in columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        
        # Calculate the Interquartile Range (IQR)
        IQR = Q3 - Q1
        
        # Determine the bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Filter the dataframe to remove outliers
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Example usage:
# Assuming `data` is your dataframe and `numerical_columns` is a list of numerical columns
numerical_columns = ['price_as_$', 'quantity', 'reviews_sum']
cleaned_data = remove_outliers(data02, numerical_columns)

# Print or save the cleaned data
print(cleaned_data)
```

```python
import re

# Function to clean product names
def clean_product_name(name):
    # Extract the primary name of the product using a regex pattern
    # Adjust the pattern as needed based on the dataset
    cleaned_name = re.sub(r"\s*\(.*?\)|\s*-\s*.*", "", name)  # Remove text within parentheses or after dashes
    return cleaned_name.strip()

# Apply the cleaning function to the product_name column
data['product_name_cleaned'] = data['product_name'].apply(clean_product_name)

# Display the cleaned product names for verification
print(data[['product_name', 'product_name_cleaned']].head())
```

```python
# Drop the product_name column
data = data.drop(columns=['product_name'])

```

```python
# Function to clean the price column
def clean_price(value):
    # Remove dollar signs, commas, and any non-numeric characters
    if pd.isna(value):
        return None
    cleaned_value = re.sub(r'[^\d.]', '', str(value))  # Retain only numbers and decimal points
    return float(cleaned_value) if cleaned_value else None

# Apply the cleaning function to the price column
data['price_as_$'] = data['price'].apply(clean_price)

# Display a sample of the original and cleaned price columns
data.head()
```

```python
# Drop the price column
data = data.drop(columns=['price'])
data.head()
```

```python
# Function to limit product name to 20 words
def limit_to_20_words(name):
    words = name.split()  # Split the name into words
    limited_words = words[:20]  # Retain only the first 40 words
    return ' '.join(limitled_words)  # Join back into a string

# Apply the function to the product_name column
data['product_names'] = data['product_name_cleaned'].apply(limit_to_20_words)

# Drop the product_name column
data = data.drop(columns=['product_name_cleaned'])

data.head()
```

```python
# Function to clean the reviews column
def clean_reviews(value):
    if pd.isna(value):
        return None  # Handle missing values
    # Extract only numeric values
    cleaned_value = re.sub(r'[^\d]', '', str(value))  # Retain only digits
    return int(cleaned_value) if cleaned_value else None

# Apply the cleaning function to the reviews column
data['reviews_sum'] = data['reviews'].apply(clean_reviews)

# Drop the review column
data = data.drop(columns=['reviews'])

# Display a sample of the original and cleaned price columns
data.head()
```

```python
from datetime import datetime
import re

# Function to clean and standardize the purchase_date column
def standardize_date(value):
    if pd.isna(value):
        return value  # Keep missing values as is
    try:
        # Attempt to parse dates with natural language
        match = re.search(r'(\b\w+\b) (\d{1,2}), (\d{4})', value)  # Extract formats like "December 6, 2024"
        if match:
            month_name, day, year = match.groups()
            month_number = datetime.strptime(month_name, "%B").month  # Convert month name to number
            standardized_date = f"{int(day):02d}-{int(month_number):02d}-{year}"  # Reformat to DD-MM-YYYY
            return standardized_date
        # Return original value if no valid date detected
        return value
    except Exception as e:
        return value  # Handle unexpected cases gracefully

# Apply the function to the purchase_date column
data['purchase_Date'] = data['purchase_date'].apply(standardize_date)

# Display a sample of the original and cleaned purchase_date columns
data.head()
```

```python
# Drop the product_name column
data = data.drop(columns=['purchase_date'])

# Display a sample of the original and cleaned purchase_date columns
data.head()
```

```python
#Function to clean the quantity column
def clean_quantity(value):
    if pd.isna(value):
        return value  # Keep missing values as is
    try:
        # Extract the first numeric value in the string
        match = re.search(r'\d+', value)
        return int(match.group()) if match else None  # Convert to integer
    except Exception as e:
        return None  # Handle unexpected cases gracefully

# Apply the cleaning function to the quantity column
data['quantity_cleaned'] = data['quantity'].apply(clean_quantity)

# Remove decimal points by converting to integer
data['quantity_cleaned'] = data['quantity_cleaned'].fillna(0).astype(int)

# Drop the product_name column
data = data.drop(columns=['quantity'])

# Display a sample of the original and cleaned quantity columns
data.head()
```

```python
#  Rename a quantity column as quantity
data.rename(columns={'quantity_cleaned': 'quantity'}, inplace=True)
```

```python
# Define the desired order of columns
desired_order = ['Customer_ID', 'product_names', 'price_as_$', 'purchase_Date', 'quantity', 'availability', 'rating', 'reviews_sum' ]  # Adjust column names as needed

# Reorder the DataFrame columns
data = data[desired_order]

# Verify the new column order
data.head()
```

### A filter function to find purchases within a specific date range or exceeding a given amount.

```python

import tkinter as tk
from tkinter import ttk, messagebox
from datetime import datetime



# Convert 'purchase_Date' to datetime format
data03['purchase_Date'] = pd.to_datetime(data03['purchase_Date'])

# Define the function to filter data
def filter_purchases(start_date, end_date, min_amount):
    try:
        # Convert string dates to datetime objects
        start_date = datetime.strptime(start_date, '%Y-%m-%d') if start_date else None
        end_date = datetime.strptime(end_date, '%Y-%m-%d') if end_date else None
    except ValueError:
        messagebox.showerror("Date Error", "Invalid date format. Use YYYY-MM-DD.")
        return None
    
    # Filter based on date range
    filtered_data = data03.copy()
    if start_date:
        filtered_data = filtered_data[filtered_data['purchase_Date'] >= start_date]
    if end_date:
        filtered_data = filtered_data[filtered_data['purchase_Date'] <= end_date]
    
    # Filter based on minimum purchase amount
    if min_amount:
        try:
            min_amount = float(min_amount)
            filtered_data = filtered_data[filtered_data['total_spent'] >= min_amount]
        except ValueError:
            messagebox.showerror("Amount Error", "Minimum amount must be a valid number.")
            return None
    
    return filtered_data

# Define the function to handle filter button click
def on_filter_click():
    start_date = start_date_entry.get()
    end_date = end_date_entry.get()
    min_amount = min_amount_entry.get()
    
    # Filter the data
    result = filter_purchases(start_date, end_date, min_amount)
    if result is None or result.empty:
        messagebox.showinfo("No Results", "No purchases match the filter criteria.")
    else:
        # Display the results in a table
        result_window = tk.Toplevel(root)
        result_window.title("Filtered Purchases")
        result_table = ttk.Treeview(result_window, columns=list(result.columns), show='headings')
        
        # Add column headers
        for col in result.columns:
            result_table.heading(col, text=col)
            result_table.column(col, width=120)
        
        # Add rows to the table
        for _, row in result.iterrows():
            result_table.insert('', tk.END, values=row.tolist())
        
        result_table.pack(fill=tk.BOTH, expand=True)
        result_window.geometry("1000x400")

# Create the main GUI window
root = tk.Tk()
root.title("Purchase Filter")
root.geometry("400x300")

# Add labels and entry fields for filtering
tk.Label(root, text="Start Date (YYYY-MM-DD):").pack(pady=5)
start_date_entry = tk.Entry(root, width=30)
start_date_entry.pack(pady=5)

tk.Label(root, text="End Date (YYYY-MM-DD):").pack(pady=5)
end_date_entry = tk.Entry(root, width=30)
end_date_entry.pack(pady=5)

tk.Label(root, text="Minimum Purchase Amount:").pack(pady=5)
min_amount_entry = tk.Entry(root, width=30)
min_amount_entry.pack(pady=5)

# Add a filter button
filter_button = tk.Button(root, text="Apply Filter", command=on_filter_click)
filter_button.pack(pady=20)

# Run the GUI loop
root.mainloop()
```

### A summary report highlighting top-performing products with additional scraped data (e.g., price trends or average ratings).

```python
import pandas as pd
import tkinter as tk
from tkinter import ttk, messagebox

# Load the dataset
data04 = pd.read_csv('data03.csv')  # Replace with your actual dataset path

# Debug: Print initial dataset preview
print("Initial Dataset Preview:")
print(data04.head())

# Clean `rating` column: Extract numeric value from strings like "4.5 out of 5"
data04['rating'] = data04['rating'].str.extract(r'(\d+\.?\d*)').astype(float)

# Clean `product_names` and `product_category`
data04['product_names'] = data04['product_names'].str.strip().str.lower()
data04['product_category'] = data04['product_category'].str.strip().str.lower()

# Convert `total_spent` and `quantity` to numeric
data04['total_spent'] = pd.to_numeric(data04['total_spent'], errors='coerce')
data04['quantity'] = pd.to_numeric(data04['quantity'], errors='coerce')

# Drop rows with critical missing values
data04 = data04.dropna(subset=['product_names', 'total_spent', 'rating', 'quantity'])

# Debug: Check cleaned dataset
print("Cleaned Dataset Preview:")
print(data04.head())
print("Dataset Info:")
print(data04.info())

# Generate a summary of top-performing products
def generate_summary():
    try:
        # Aggregate metrics: total revenue, average rating, total quantity sold
        summary = data04.groupby('product_names').agg(
            total_revenue=('total_spent', 'sum'),
            average_rating=('rating', 'mean'),
            total_quantity=('quantity', 'sum')
        ).sort_values(by='total_revenue', ascending=False).head(10)  # Top 10 products

        print("Summary Data:")  # Debug: Check the summary data
        print(summary)

        return summary
    except Exception as e:
        print(f"Error in generate_summary: {e}")
        return pd.DataFrame()  # Return an empty DataFrame if an error occurs

# Define the function to display the summary report
def display_summary_report():
    # Generate the summary
    summary = generate_summary()
    
    if summary.empty:
        messagebox.showinfo("No Data", "No data available for summary.")
        return
    
    # Create a new window to display the summary
    summary_window = tk.Toplevel(root)
    summary_window.title("Top-Performing Products Summary")
    summary_window.geometry("800x400")
    
    # Create a Treeview widget for the summary table
    columns = ['Product Name', 'Total Revenue', 'Average Rating', 'Total Quantity']
    summary_table = ttk.Treeview(summary_window, columns=columns, show='headings')
    
    # Configure the column headers
    for col in columns:
        summary_table.heading(col, text=col)
        summary_table.column(col, width=200)
    
    # Insert rows into the table
    for product, row in summary.iterrows():
        summary_table.insert('', tk.END, values=[
            product.capitalize(),
            f"${row['total_revenue']:.2f}",
            f"{row['average_rating']:.2f}",
            int(row['total_quantity'])
        ])
    
    summary_table.pack(fill=tk.BOTH, expand=True)

# Create the main GUI window
root = tk.Tk()
root.title("Summary Report: Top-Performing Products")
root.geometry("400x200")

# Add a button to generate the summary report
generate_button = tk.Button(root, text="Generate Summary Report", command=display_summary_report)
generate_button.pack(pady=50)

# Run the GUI loop
root.mainloop()
```
